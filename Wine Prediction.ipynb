```python
import pandas as pd

dataset_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
data = pd.read_csv(dataset_url, sep=";")# The sep=";" removes the comas separation in our dataset.
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# tools for cross- Validation
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

# Tools to evaluate our model later
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Tools to persist our model for future use(Like Python Pickle)
import joblib
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
# Selecting our target variables and our features
y = data.quality

X = data.drop("quality", axis =1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123, stratify=y)
# random_state is the speed, stratify is to make sure that train test and test set look similar. Stratify your sample by the target variable.
```

```python
# A scaler for Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# define data preprocessing tools
pipeline = make_pipeline(
           StandardScaler(), # for scaling our features
           RandomForestRegressor(n_estimators = 100, random_state =123) # Model Fitting Step
)
```

```python

print(pipeline.get_params())
```

```python
# defining hyperparameters to tune
hyperparameters = {'randomforestregressor__max_features': ['auto', 'sqr', 'log2'],
                   'randomforestregressor__max_depth': [None, 5, 3, 1]}
```

```python
# Tune model through Cross Validation(10 K-folds) Pipeline
clf = GridSearchCV(pipeline, hyperparameters, cv =10 )
```

```python
# Tune and fit our model
clf.fit(X_train, y_train)
```

```python
# Checking for the best Hyperparameters
print(clf.best_params_)
```

```python
# Checking for the entire refiting of our training set
#  **GridSearchCV does it automatically** #

print(clf.refit)
```

```python
# Evaluate Model Performance on Test Data.
y_pred = clf.predict(X_test)

```

```python
print(f'R2 Score : {r2_score(y_test,y_pred)}')
print(f'Mean Absolute Error : {mean_absolute_error(y_test,y_pred)}')
```

```python
# Saving the Model For Future Use
joblib.dump(clf, 'WinePrediction.pkl')

# To load: clf2 = joblib.load('WinePrediction.pkl')
```
